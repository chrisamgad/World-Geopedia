```python
#Get Capital Cities and Countries
import requests
from bs4 import BeautifulSoup
import re
  
URL = "https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_continent"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
table = soup.findAll('table')


AfricaCountries=[]
AsiaCountries=[]
EuropeCountries=[]
NorthAmericaCountries=[]
OceaniaCountries=[]
SouthAmericaCountries=[]
AntarticaCountries=[]

All_Countries=[]
CorrespondingContinents=[]

rowsAfrica = table[1].find_all('tr') 
for row in rowsAfrica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AfricaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Africa")

rowsAsia = table[2].find_all('tr') 
for row in rowsAsia:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                AsiaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Asia")
                
rowsEurope = table[3].find_all('tr') 
for row in rowsEurope:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                EuropeCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Europe")

rowsNorthAmerica = table[4].find_all('tr') 
for row in rowsNorthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                NorthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("NorthAmerica")
                
rowsOceania= table[5].find_all('tr') 
for row in rowsOceania:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                OceaniaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Oceania")

rowsSouthAmerica= table[6].find_all('tr')    

for row in rowsSouthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                SouthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("SouthAmerica")

rowsAntartica= table[7].find_all('tr') 
for row in rowsAntartica:
    cells=row.find_all('td')
    for cell in cells:
        i_tags=cell.find_all('i')
        for i_tag in i_tags:
            a_country_tags=i_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AntarticaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Antartica")
                
#for a in AfricaCountries:
#    print(a['title'] + "\n" + a['href'])
for idx, country in enumerate(All_Countries):
    print(country['title'] + " in "+ CorrespondingContinents[idx]+"\n" + country['href'])
print("N_countires = "+str(len(All_Countries)))
print("N_continents = "+str(len(All_Countries)))

# for a in All_Countries:
#     print(a['title'] + "\n" + a['href'])
```

```python
#GetPresidentBirthDate
import requests
from bs4 import BeautifulSoup
import re

def GetPresidentBOD (President):  
    if(President==None):
        print("None President")
        return "Null"
    else:
        print(President.text)
        print(President['href'])

    url="https://en.wikipedia.org"+President['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 

    table=soup.findAll('table',class_="infobox vcard")
    if(not table): #If still no table
        print("Null")
        return "Null"
        


    FoundbDay=False
    BirthDate=""
    Bday=table[0].findAll('span',class_='bday')
    if(Bday):
        BirthDate=Bday[0].text
        print(Bday[0].text)
        return BirthDate
    else:
        print("Null")
        return "Null"
        
    
PresidentBODs=[]
for president in Presidents:
    PresiBOD=GetPresidentBOD(president)
    PresidentBODs.append(PresiBOD)
    
```

```python
def GetPresidentOfficeAssumptionDate(President):
    if(President==None):
        print("None President")
        return "Null"
    else:
        print(President.text)
        print(President['href'])

    url="https://en.wikipedia.org"+President['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 

    table=soup.findAll('table',class_="infobox vcard")
    if(not table): #If still no table
        print("Null")
        return "Null"

    AssumedDate=""
    FoundAssuumedDate=False
    rows=table[0].findAll('tr')
    for row in rows:
        b_tag_assumed_office=row.findAll('b', string="Assumed office")
        if(b_tag_assumed_office):
            AssumedOffice_td_Tag=b_tag_assumed_office[0].find_parent('td')
            if(AssumedOffice_td_Tag):
                AssumedDate=AssumedOffice_td_Tag.text.split('[')[0].split('(')[0].split('Assumed office ')[1].split('Acting')[0]
                FoundAssuumedDate=True
                print(AssumedDate)
                break
        if(FoundAssuumedDate):
            break

    if(not FoundAssuumedDate):
        print("Null")
        return "Null"
    return AssumedDate
PresidentAssumedDates=[]
for president in Presidents:
    PresiAssumedDate=GetPresidentOfficeAssumptionDate(president)
    #print(PresiAssumedDate)
    PresidentAssumedDates.append(PresiAssumedDate)
    
```

```python
def GetPoliticalParty(President):

    if(President==None):
        print("None President")
        return "Null"
    else:
        print(President.text)
        print(President['href'])

    url="https://en.wikipedia.org"+President['href']#/wiki/Nicolás_Maduro"
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table=soup.findAll('table',class_="infobox vcard")

    if(not table): #If still no table
        print("Null")
        return "Null"

    PolicialParty=""
    FoundPolicialParty=False
    rows=table[0].findAll('tr')
    for row in rows:
        th_tags=row.findAll('th', string="Political party")
        if(th_tags):
            #print(th_tags[0].find_next_sibling('td'))
            PoliticalParty_a=th_tags[0].find_next_sibling('td').a
            if(PoliticalParty_a):
                PolicialParty=PoliticalParty_a.text
                FoundPolicialParty=True
                print(PolicialParty)
                break
    if(not FoundPolicialParty):
        print("Null")
        return "Null"
    return PolicialParty

PresidentPoliticalParty=[]
for president in Presidents:
    PresiPoliticalParty=GetPoliticalParty(president)
    PresidentPoliticalParty.append(PresiPoliticalParty)
```

```python
#country Names
CountryNames=[]
for country in All_Countries:
        CountryNames.append(country['title'])

print(len(CountryNames))  
```

```python

PresidentsNames_Copy=PresidentsNames.copy()
#print(PresidentsNames_Copy)
PresidentPoliticalParty_Copy=PresidentPoliticalParty.copy()
PresidentAssumedDates_Copy=PresidentAssumedDates.copy()
PresidentBODs_Copy=PresidentBODs.copy()
CountryNames_Copy=CountryNames.copy()

PresidentsNamesWithoutDup=[]
PresidentPoliticalPartyWithoutDup=[]
PresidentAssumedDatesWithoutDup=[]
PresidentBODsWithoutDup=[]
CountryNamesWithoutDup=[]

print(len(PresidentsNames_Copy))
print(len(PresidentPoliticalParty_Copy))
print(len(PresidentAssumedDates_Copy))
print(len(PresidentBODs_Copy))
print(len(CountryNames_Copy))
```

```python
#REMOVE ANY UNSPECIFIED PRESIDENT
res=[]

for i,country in enumerate(CountryNames_Copy):
    if country not in res :
        res.append(country)
        PresidentsNamesWithoutDup.append(PresidentsNames_Copy[i])
        PresidentPoliticalPartyWithoutDup.append(PresidentPoliticalParty_Copy[i])
        PresidentAssumedDatesWithoutDup.append(PresidentAssumedDates_Copy[i])
        PresidentBODsWithoutDup.append(PresidentBODs_Copy[i])
        CountryNamesWithoutDup.append(country)
                
print("Done removing duplicates")
```

```python
print(len(PresidentsNamesWithoutDup))
print(len(PresidentPoliticalPartyWithoutDup))
print(len(PresidentAssumedDatesWithoutDup))
print(len(PresidentBODsWithoutDup))
print(len(CountryNamesWithoutDup))
#print(CountryNamesWithoutDup)
```

```python
#PreProcessing stage
FinalPresidentNames=PresidentsNamesWithoutDup
FinalPoliticalParty=[]
FinalPresidentAssumedDates=[]
FinalPresidentBODs=[]
FinalCountryNames=[]

for countryname in CountryNamesWithoutDup:
        FinalCountryNames.append(countryname)
        
for politicalParty in PresidentPoliticalPartyWithoutDup:
    if("Null" in politicalParty):
        FinalPoliticalParty.append(None)
    else:
        FinalPoliticalParty.append(politicalParty)

for assumed_date in PresidentAssumedDatesWithoutDup:
    if(assumed_date==None or "Null" in assumed_date ):
        FinalPresidentAssumedDates.append(None)
    else:
        FinalPresidentAssumedDates.append(assumed_date)
        
for BOD in PresidentBODsWithoutDup:
    if("Null" in BOD):
        FinalPresidentBODs.append(None)
    else:
        FinalPresidentBODs.append(BOD)

```

```python
print(len(FinalPresidentNames))
print(len(FinalPoliticalParty))
print(len(FinalPresidentAssumedDates))
print(len(FinalPresidentBODs))
print(len(FinalCountryNames))
#print(FinalPresidentAssumedDates)
```

```python
presidents={
    "CountryName":FinalCountryNames,
    "PresidentName":FinalPresidentNames,
    "PoliticalParty":FinalPoliticalParty,
    "BOD":FinalPresidentBODs,
    "AssumptionDate":FinalPresidentAssumedDates
}
```

```python
import pandas as pd
df = pd.DataFrame(presidents)
df.to_csv("Presidents.csv")
```
