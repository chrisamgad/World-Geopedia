```python
import pandas as pd
import re

#Get Capital Cities and Countries
import requests
from bs4 import BeautifulSoup
import re
  
URL = "https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_continent"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
table = soup.findAll('table')


AfricaCountries=[]
AsiaCountries=[]
EuropeCountries=[]
NorthAmericaCountries=[]
OceaniaCountries=[]
SouthAmericaCountries=[]
AntarticaCountries=[]

All_Countries=[]
CorrespondingContinents=[]

rowsAfrica = table[1].find_all('tr') 
for row in rowsAfrica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AfricaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Africa")

rowsAsia = table[2].find_all('tr') 
for row in rowsAsia:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                AsiaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Asia")
                
rowsEurope = table[3].find_all('tr') 
for row in rowsEurope:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                EuropeCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Europe")

rowsNorthAmerica = table[4].find_all('tr') 
for row in rowsNorthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                NorthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("NorthAmerica")
                
rowsOceania= table[5].find_all('tr') 
for row in rowsOceania:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                OceaniaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Oceania")

rowsSouthAmerica= table[6].find_all('tr')    

for row in rowsSouthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                SouthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("SouthAmerica")

rowsAntartica= table[7].find_all('tr') 
for row in rowsAntartica:
    cells=row.find_all('td')
    for cell in cells:
        i_tags=cell.find_all('i')
        for i_tag in i_tags:
            a_country_tags=i_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AntarticaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Antartica")
                
#for a in AfricaCountries:
#    print(a['title'] + "\n" + a['href'])
for idx, country in enumerate(All_Countries):
    print(country['title'] + " in "+ CorrespondingContinents[idx]+"\n" + country['href'])
print("N_countires = "+str(len(All_Countries)))
print("N_continents = "+str(len(All_Countries)))

# for a in All_Countries:
#     print(a['title'] + "\n" + a['href'])
```

```python
#Tokelau has "None" in the value of Capital
def GetAllCapitalCities(AllCountries):
    CapitalCities=[]#list contains cap city
    CountriesandtheirCapitals=[]#list of lists (each index contains couuntry and corresponding city)
    
    for country in AllCountries:      
        print("\n")
        print(country['title'])
        #temp_countryAndCapital=[]
        #temp_countryAndCapital.append(country)
        
        r = requests.get("https://en.wikipedia.org"+country['href'])
        soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
        table = soup.findAll('table',class_="infobox ib-country vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-pol-div vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-islands vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-settlement vcard")
                    if(not table):
                        table=soup.findAll('table',class_="infobox ib-disputed-islands")
        rows=table[0].findAll('tr')
        #print(rows)
        
        CapCity=None#initially
        if (country['title']=="Yemen"): #Corner Case
            for row in rows:
                #CapCity=None
                a_s=row.findAll('a')
                for a in a_s:
                    if(a['href']=="/wiki/Sana%27a"):
                        CapCity=a;
                        print(CapCity['title'])
                        print(CapCity['href'])
                        CapitalCities.append(CapCity) 
                        break;
                if(CapCity):
                    break;
        elif(country['title']=="Tokelau"):#Corner Case
            CapCity=None
        elif(country['href']=="/wiki/Sahrawi_Arab_Democratic_Republic"):#Corner Case
            for row in rows:
                #CapCity=None
                a_s=row.findAll('a')
                for a in a_s:
                    if(a['href']=="/wiki/Laayoune"):
                        CapCity=a;
                        print(CapCity['title'])
                        print(CapCity['href'])
                        CapitalCities.append(CapCity)  
                        break;
                if(CapCity):
                    break;
        elif(rows):
            CapCity=None #initally none until finds one
            for row in rows:
                cells=row.findAll('th')
                if(cells):
                    for cell in cells:
                        if("Capital" in cell.text):
                            #print("test")
                            test=cell.find_next_siblings('td')[0]
                            #a_test=test.findAll('a', class_=lambda x: "image") #class != image because a small number of countries have more than a single link
                            a_test=test.findAll('a')

                            for a in a_test:                
                                if(a.has_attr("class")):
                                    #print(a["class"])
                                    if(not(a["class"][0]=="external" or a["class"][0]=="image" or a['title']=="Geographic coordinate system")):#Corner Cases
                                        CapCity=a
                                        CapitalCities.append(CapCity)
                                        #temp_countryAndCapital.append(CapCity)
                                        print(CapCity['title'])
                                        print(CapCity['href'])
                                        break;
                                else:
                                    CapCity=a
                                    CapitalCities.append(CapCity)
                                    #temp_countryAndCapital.append(CapCity)
                                    print(CapCity['title'])
                                    print(CapCity['href'])
                                    break;                                  
        if CapCity==None:
            CapitalCities.append(CapCity) 
            #temp_countryAndCapital.append(CapCity)
            print("NULL CAPITAL CITY")
        #CountriesandtheirCapitals.append(temp_countryAndCapital)
        #print(CountriesandtheirCapitals)
    return CapitalCities

CapitalCities=GetAllCapitalCities(All_Countries) #Takes in Array of All countries

print("DONEEE")
```

```python
import requests
from bs4 import BeautifulSoup
import re

def RemoveSpecialChars(str1):
    alphanumeric = ""
    for character in str1:
        if character.isalnum():
            alphanumeric += character
    return alphanumeric

def GetCityCoordinates(City):
    if(City==None):
        print("None City")
        return "Null"
    else:
        print(City.text)
        print(City['href'])
    url="https://en.wikipedia.org"+City['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-islands vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-pol-div vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
                    if(not table):
                        table=soup.findAll('table',class_="infobox-full-data")

    if(not table): #If still no table
        print("Null")
        return "Null"
    FoundCoordinates=False
    CityCoordinates=""
    longitude=""
    latitude=""
    Coordinates=""
    rows=table[0].findAll('tr')
    for row in rows:
        span_tags=row.findAll('span',class_="latitude")
        if(span_tags):
            #print(span_tags[0].text)
            latitude=span_tags[0].text
            
        span_tags=row.findAll('span',class_="longitude")
        if(span_tags):
            #print(span_tags[0].text)
            longitude=span_tags[0].text

    if(longitude != ""):
        Coordinates=Coordinates+longitude
    if(latitude!=""):
        Coordinates=Coordinates+latitude
    #print(RemoveSpecialChars(Coordinates))
        Coordinates=Coordinates.replace('°',"'").replace('′','`').replace("′′",'``')
        print(Coordinates)
    if(Coordinates==""):
        return "Null"
    
    return Coordinates


CityCoordinates=[]
for city in CapitalCities:
    coordinates=GetCityCoordinates(city)
    CityCoordinates.append(coordinates)
    
```

```python
def GetCityPopulation(url):
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-islands vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-pol-div vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
                    if(not table):
                        table=soup.findAll('table',class_="infobox-full-data")
    
    if(not table): #If still no table
        print("Null")
        return "Null"
    rows=table[0].findAll('tr')    
    FoundCityPopulation=False
    population=""
    for row in rows:
        th_tags=row.findAll('th')
        Metro_atag=row.findAll('a', href="/wiki/Metropolitan_area")
        Urban_atag=row.findAll('a', href="/wiki/Urban_area")
        if(Metro_atag):
            Metro_a_thtag=Metro_atag[0].find_parent('th').find_next_sibling('td')
            if(Metro_a_thtag):
                population=Metro_a_thtag.text
                population=population.split('[')[0]
                population=population.split('(')[0].split('km')[0].split('/')[0]
                population=population.strip().replace(',','')
                FoundCityPopulation=True
                print(population)
                break;
        else:
            for th in th_tags:
                #print(th)
                if("Population" in th.text):
                    population=row.find_next_sibling("tr").td.text
                    population=population.split('[')[0]
                    population=population.split('(')[0].split('km')[0].split('/')[0]
                    population=population.strip().replace(',','')
                    FoundCityPopulation=True
                    print(population)
                    break;
                    
                if(FoundCityPopulation):
                    break

            if(FoundCityPopulation):
                break
    if("South" in population):
        population=population.split('South: ')[1]
        return population
    elif(population[0] and "million" not in population):
        try:
            if(population[0].isdigit()==False):
                population="Null"
                print("Null")
            elif(population[1]):
                if(population[1].isdigit()==False):
                    population="Null"
                    print("Null")
        except:
            pass
    if(FoundCityPopulation==False):
        population="Null"
        print("Null")

    #print(population)
    return population
CityPopulation=[]
for city in CapitalCities:
    if(city==None):
        CityPopulation.append("Null")
    else:
        citylink="https://en.wikipedia.org"+city['href'] 
        CityPopulation.append(GetCityPopulation(citylink))
```

```python
#Get City Governor
import requests
from bs4 import BeautifulSoup
import re

def GetCityGovernor(City):
    if(City==None):
        print("None City")
        return "Null"
    else:
        print(City.text)
        print(City['href'])
    
    url="https://en.wikipedia.org"+City["href"]
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-islands vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-pol-div vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
                    if(not table):
                        table=soup.findAll('table',class_="infobox-full-data")
    if(not table): #If still no table
        print("Null")
        return "Null"

    FoundGovernor=False
    Governor=""
    rows=table[0].findAll('tr') 
    for row in rows:
        th_tags=row.findAll('th')
        for th in th_tags:
           # print(th)
            if(th):
                if("Government" in th):
                    #if("Governor" in th.text):  
                    GovernorKey=row.find_next_sibling('tr')
                    #print(GovernorKey)
                    if("Governor" in GovernorKey.th.text):
                        Governor=GovernorKey.td.text
                        Governor=Governor.split('[')[0]
                        Governor=Governor.split('(')[0]
                        print(Governor)
                        FoundGovernor=True
                        break
                    elif("Mayor" in GovernorKey.th.text):
                        Governor=GovernorKey.td.text
                        Governor=Governor.split('[')[0]
                        Governor=Governor.split('(')[0]
                        print(Governor)
                        FoundGovernor=True
                        break
                    else:
                        GovernorKey=GovernorKey.find_next_sibling('tr')
                        #print(GovernorKey)
                        if(GovernorKey.th):
                            #print(GovernorKey.th)
                            if("Governor" in GovernorKey.th.text):
                                Governor=GovernorKey.td.text
                                Governor=Governor.split('[')[0]
                                Governor=Governor.split('(')[0]
                                print(Governor)
                                FoundGovernor=True
                                break
                            elif("Mayor" in GovernorKey.th.text):
                                Governor=GovernorKey.td.text
                                Governor=Governor.split('[')[0]
                                Governor=Governor.split('(')[0]
                                print(Governor)
                                FoundGovernor=True
                                break
        if(FoundGovernor):
            break

    if(not FoundGovernor):
        print("Null")
        return "Null"
    
    return Governor

CityGovernors=[]
for city in CapitalCities:
    Governor=GetCityGovernor(city)
    CityGovernors.append(Governor)
```

```python
import requests
from bs4 import BeautifulSoup
import re
#Get Capital Cities Area
def GetCityArea(City):

    if(City==None):
        print("None City")
        return "Null"
    else:
        print(City.text)
        print(City['href'])
        
    url="https://en.wikipedia.org"+City["href"]
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-islands vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-pol-div vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
                    if(not table):
                        table=soup.findAll('table',class_="infobox-full-data")

    if(not table): #If still no table
        print("Null")
        return "Null"
    FoundArea=False
    CityArea=""
    rows=table[0].findAll('tr') 
    for row in rows:
        th_tags=row.findAll('th')
        for th in th_tags:
            if("Area" in th.text):
                trContainingArea=th.find_parent('tr').find_next_sibling('tr')
                if(trContainingArea):      
                    CityArea=trContainingArea.td
                    if(CityArea):
                        if("km" in CityArea.text):
                            CityArea=CityArea.text
                            CityArea=CityArea.split('\xa0')[0]
                            CityArea=CityArea.strip().replace(',','')
                            print(CityArea)
                            FoundArea=True
                            break;
            
            if(FoundArea):
                break;        
        if(FoundArea): 
            break;    
    if(FoundArea==False):
        CityArea="Null"
        print(CityArea)
    return CityArea

CityAreas=[]
for city in CapitalCities:
    Area=GetCityArea(city)
    CityAreas.append(Area)
```

```python
FinalCountries=CountryNamesWithoutDup
FinalCapCities=CapitalCityNamesWithoutDup
FinalArea=[]
FinalCityGovernors=[]
FinalCityCoordinates=[]
FinalCityPopulation=[]

for i,population in enumerate(CityPopulationWithoutDup):
    if("Null" in population):
        FinalCityPopulation.append(None)
    else:
        population=population.split('/')[0].split('k')[0].split('%')[0].strip()
        FinalCityPopulation.append(population)
    
for i,coordinates in enumerate(CityCoordinatesWithoutDup):
    if("Null" in coordinates):
        FinalCityCoordinates.append(None)
    else:
        FinalCityCoordinates.append(str(coordinates))

for i,governor in enumerate(CityGovernorsWithoutDup):
    #print(governor)
    if("Null" in coordinates):
        FinalCityGovernors.append(None)
    else:
        FinalCityGovernors.append(governor)

for i, area in enumerate(CityAreasWithoutDup):
    if("Null" in area):
        FinalArea.append(None)
    else:
        area=float(area)
        FinalArea.append(area)
```
