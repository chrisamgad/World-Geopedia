```python
import pandas as pd
import re

#Get Capital Cities and Countries
import requests
from bs4 import BeautifulSoup
import re
  
URL = "https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_continent"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
table = soup.findAll('table')


AfricaCountries=[]
AsiaCountries=[]
EuropeCountries=[]
NorthAmericaCountries=[]
OceaniaCountries=[]
SouthAmericaCountries=[]
AntarticaCountries=[]

All_Countries=[]
CorrespondingContinents=[]

rowsAfrica = table[1].find_all('tr') 
for row in rowsAfrica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AfricaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Africa")

rowsAsia = table[2].find_all('tr') 
for row in rowsAsia:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                AsiaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Asia")
                
rowsEurope = table[3].find_all('tr') 
for row in rowsEurope:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                EuropeCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Europe")

rowsNorthAmerica = table[4].find_all('tr') 
for row in rowsNorthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
               # print(a)
                NorthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("NorthAmerica")
                
rowsOceania= table[5].find_all('tr') 
for row in rowsOceania:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                OceaniaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Oceania")

rowsSouthAmerica= table[6].find_all('tr')    

for row in rowsSouthAmerica:
    cells = row.find_all('td')
    for cell in cells:
        b_tags=cell.find_all('b')
        for b_tag in b_tags:
            a_country_tags=b_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                SouthAmericaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("SouthAmerica")

rowsAntartica= table[7].find_all('tr') 
for row in rowsAntartica:
    cells=row.find_all('td')
    for cell in cells:
        i_tags=cell.find_all('i')
        for i_tag in i_tags:
            a_country_tags=i_tag.find_all('a')
            for a in a_country_tags:
                #print(a)
                AntarticaCountries.append(a)
                All_Countries.append(a)
                CorrespondingContinents.append("Antartica")
                
#for a in AfricaCountries:
#    print(a['title'] + "\n" + a['href'])
for idx, country in enumerate(All_Countries):
    print(country['title'] + " in "+ CorrespondingContinents[idx]+"\n" + country['href'])
print("N_countires = "+str(len(All_Countries)))
print("N_continents = "+str(len(All_Countries)))

# for a in All_Countries:
#     print(a['title'] + "\n" + a['href'])
```

```python
from datetime import date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

import numpy as np
from datetime import date
from faker import Faker
import pandas as pd

fake = Faker()
gender = np.random.choice(["M", "F"], p=[0.5, 0.5])
usernames=[]
emails=[]
genders=[]
birthdates=[]
ages=[]

Users={}

for i in range(0,100):
    usernames.append(fake.user_name())
    emails.append(fake.email())
    genders.append(np.random.choice(["M", "F"], p=[0.5, 0.5]))
    BOD=fake.date_of_birth()
    birthdates.append(BOD)
    ages.append(calculate_age(BOD))
    

Users={
    "Username": usernames,
    "Email":emails,
    "Gender":genders,
    "BirthDate":birthdates,
    "Age":ages
}

#Create User Table
df = pd.DataFrame(Users)
df.to_csv("Users.csv")
```
