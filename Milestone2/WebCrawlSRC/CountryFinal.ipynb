```python
def readCountry(link):
    try:
        country = pd.read_html(link, attrs={"class":"infobox ib-country vcard" })
    except:
        pass
    try:
        country = pd.read_html(link, attrs={"class": "infobox ib-islands vcard"})
    except:
        pass
    try:
        country = pd.read_html(link, attrs={"class":"infobox ib-pol-div vcard" })
    except:
        pass
    try:
        country = pd.read_html(link, attrs={"class":"infobox ib-settlement vcard" })
    except:
        pass
    try:
        country = pd.read_html(link, attrs={"class":"infobox ib-disputed-islands" })
    except:
        pass
    x = len(country)
    print(x)
    country = country[0].dropna().reset_index()
    country = country.rename(columns={country.columns[1]: 'Variable', country.columns[2]: 'Value'})
    country = country.drop("index", axis=1)
    
    return country
```

```python
def GetOfficialLanguage(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-pol-div vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-islands vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")

    FoundOfficialLanguage=False
    OfficialLanguage=""
    rows=table[0].findAll('tr')
    for row in rows:
        th_tags=row.findAll('th')
        for th in th_tags:
            #print(th.text)
            if(("Official"in th.text) & (("languages"in th.text) or ("language"in th.text))):#and keyword due to special character space in between sometimes
                OfficialLanguage=th.find_next_sibling('td').ul
                if(OfficialLanguage):
                    #print(OfficialLanguage)
                    OfficialLanguage=OfficialLanguage.li#.text
                    if(OfficialLanguage.a):
                        OfficialLanguage=OfficialLanguage.a.text
                    else:
                        OfficialLanguage=OfficialLanguage.text
                        #print(OfficialLanguage)
                        
                    OfficialLanguage=OfficialLanguage.split('[')[0]
                    FoundOfficialLanguage=True
                    
                    break
                else:
                    OfficialLanguage=th.find_next_sibling('td').a
                    if(OfficialLanguage):
                        OfficialLanguage=OfficialLanguage.text
                        OfficialLanguage=OfficialLanguage.split('[')[0]
                        FoundOfficialLanguage=True
                        #print("tesd")
                        break

            if(FoundOfficialLanguage):
                break;
        if(FoundOfficialLanguage):
            break

    if OfficialLanguage=="" or FoundOfficialLanguage==False :
        OfficialLanguage="Null"
    # print(OfficialLanguage)
    return OfficialLanguage

print(GetOfficialLanguage("https://en.wikipedia.org/wiki/Lebanon"))
```

```python
def GetArea(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-pol-div vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-islands vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
    Area=""
    rows=table[0].findAll('tr')
    FoundArea=False
    for row in rows:
        a=row.findAll('a')
        #print(row)
        if(a):
            if("Area" in a[0].text):
               # print("sdfs")
                Area=row.find_next_sibling('tr').td.text
                Area=Area.split('[')[0]
                Area=Area.split(' ')[0]
                Area=Area.split('\xa0')[0]
                FoundArea=True;
                break;
        else:
            th_tags=row.findAll('th')
            for th in th_tags:
                #print(row)
                if("Area" in th):
                    Area=row.find_next_sibling('tr').td.text
                    Area=Area.split('[')[0]
                    Area=Area.split(' ')[0]
                    Area=Area.split('\xa0')[0]
                    FoundArea=True;
                    break;
                    
            if(FoundArea):  
                break;
    if(Area==""):
        Area="Null"
    
    return Area.strip().replace(',','')
```

```python
def GetPPPGDP(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-pol-div vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-islands vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
    rows=table[0].findAll('tr')
    PPPGDP=""
    FoundPPPGDP=False
    for row in rows:
        PPP_tag= row.find_all('a', href="/wiki/Purchasing_power_parity")
        if(PPP_tag):#if found tag
            PPPGDP=PPP_tag[0].find_parent('tr').find_next_sibling('tr').td.text
            PPPGDP=PPPGDP.split('[')[0]
            PPPGDP=PPPGDP.split(' (')[0]
            FoundPPPGDP=True
            break;
        else:
            th_tags=row.findAll('th')
            for th in th_tags:
                if("PPP" in th.text):
                    print(th.text)
                    
                    PPPGDP=th.find_parent('tr').find_next_sibling('tr').td.text
                    PPPGDP=PPPGDP.text.split('[')[0]
                    PPPGDP=PPPGDP.text.split(' (')[0]
                    FoundPPPGDP=True
                    break;
                    
            if(FoundPPPGDP):
                break
    if(PPPGDP==""):
        PPPGDP="Null"

    return PPPGDP
```

```python
def GetNominalGDP(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-pol-div vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-islands vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")
    rows=table[0].findAll('tr')
    FoundNominal=False
    NominalGDP=""
    for row in rows:
        GDP= row.find_all('a', href="/wiki/Gross_domestic_product")
        if(GDP):
            NominalGDP_tag=GDP[0].find_next_sibling('span')
            if(not NominalGDP_tag):
                NominalGDP_tag=GDP[0].find_next_sibling('a')
            #print(NominalGDP_tag)
            if(NominalGDP_tag):
                if("nominal" in NominalGDP_tag.text):
                    NominalGDPText=NominalGDP_tag.find_parent('tr').find_next_sibling('tr').td.text
                    NominalGDPText=NominalGDPText.split('[')[0]
                    NominalGDPText=NominalGDPText.split('(')[0]
                    NominalGDP=NominalGDPText
                    FoundNominal=True
                    break;
            else:
                print("didn;t detect nominal gdp")
                
        else:
            th_tags=row.findAll('th')
            for th in th_tags:
                if("Nominal" in th.text or "nominal" in th.text ):
                    #print(th.text)
                    NominalGDPText=th.find_parent('tr').find_next_sibling('tr').td.text
                    NominalGDPText=NominalGDPText.split('[')[0]
                    NominalGDPText=NominalGDPText.split('(')[0]
                    NominalGDP=NominalGDPText
                    #print(NominalGDP)
                    FoundNominal=True
                    break;
                    
            if(FoundNominal):
                break
                
    if(NominalGDP==""):
        NominalGDP="Null"
    
    return NominalGDP
```

```python
def GetCountryPopulation(url):
    href_splited=url.split("https://en.wikipedia.org")[1]
    #print(href_splited)
    if(href_splited =="/wiki/Heard_Island_and_McDonald_Islands"):
        return "Null"
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
    table = soup.findAll('table',class_="infobox ib-country vcard")
    if(not table):
        table=soup.findAll('table',class_="infobox ib-pol-div vcard")
        if(not table):
            table=soup.findAll('table',class_="infobox ib-islands vcard")
            if(not table):
                table=soup.findAll('table',class_="infobox ib-settlement vcard")
                if(not table):
                    table=soup.findAll('table',class_="infobox ib-disputed-islands")

    FoundPopulation=False

    rows=table[0].findAll('tr')
    for row in rows:
        a_tags=row.findAll('a',string="Population")
        #print(row)
        if(a_tags):
            pop_text=a_tags[0].find_parent('tr').find_next_sibling('tr').td.text
            pop_text=pop_text.split('[')[0]
            pop_text=pop_text.split('(')[0]
            pop_text=pop_text.strip().replace(',','')
            FoundPopulation=True
            break;
            #print("test")
        else:
            th_tags=row.findAll('th')
            for th in th_tags:
                #print(th)
                if("Population" in th):                 
                    pop_text=th.find_parent('tr')
                    parent_tr=th.find_parent('tr')
                    if(not pop_text):
                        return "0"
                    else:
                        pop_text=pop_text.find_next_sibling('tr')
                        if(pop_text):
                            if(pop_text.td):
                                pop_text= pop_text.td.text
                                pop_text=pop_text.split('[')[0]
                                pop_text=pop_text.split('(')[0]
                                pop_text=pop_text.strip().replace(',','')
                                FoundPopulation=True
                                break
                            else:
                                if(parent_tr.td):
                                    pop_text= parent_tr.td.text
                                    pop_text=pop_text.split('[')[0]
                                    pop_text=pop_text.split('(')[0]
                                    pop_text=pop_text.strip().replace(',','')
                                    FoundPopulation=True
                                    break
                        else:
                            return "0"
                else:
                    pop_text="0"


            if(FoundPopulation):
                break;
        if(FoundPopulation):
            break;

    if(FoundPopulation==False):
        pop_text="Null"
    return pop_text
```

```python
def RemoveSpecialChars(str1):
    alphanumeric = ""
    for character in str1:
        if character.isalnum():
            alphanumeric += character
    return alphanumeric
```
